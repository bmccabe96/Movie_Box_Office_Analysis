```python
import pandas as pd
#link to IMDB datset can be found on the following line 
#https://github.com/Scogs25/Box_office_Analysis/tree/master/zippedData
```

```python
Movies_mojo=pd.read_csv('Box_Office_Mojo_Domestic_Box_Office_Chart.csv')
Movies_mojo.head()
```

```python
Movies_imdb=pd.read_csv('tmdb.movies.csv')
Movies_imdb.head()
```

```python
Movies_imdb['title_lowercase']=Movies_imdb['original_title'].apply(lambda x:x.lower().replace(' ',''))
```

```python
Movies_mojo['title_lowercase']=Movies_mojo['Title'].apply(lambda x:x.lower().replace(' ',''))
```

```python
df=Movies_mojo.merge(Movies_imdb,how='left',on='title_lowercase')
```

```python
df.drop(axis=1,columns="Unnamed: 0_x",inplace=True)
```

```python
df.to_csv('Mojo_Plus_IMDB_Data.csv')
```
