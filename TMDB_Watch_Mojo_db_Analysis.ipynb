# Analysis on the significance of the day of the week in terms of opening weekend box office sales

## How about the day of the week? Can the day of the week play a major role in box office success?

```python
## ## The key to the code below is to use datetime function which is built into pandas. 
##Using the weekday function we will determine the day of the week for each movie
##and then using group by we will see how much the day of the week matters.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt #Importing necessary Libraries
%matplotlib inline
plt.style.use('seaborn')
```

```python
df=pd.read_csv('Mojo_Plus_IMDB_Data.csv') #bringing in the dataframe
df.head()
```

```python
df.Date.head() #checking the format of the date column to see how to manipulate it
```

```python
df["Date_dt"]=df.Date.apply(lambda x:pd.to_datetime(x))
df['Date_dt'].head() #now we a datetime object, so we can use dt methods.
```

```python
df['Date_dt'].dt.dayofweek.head(20) #Monday=0 to Sunday=6
```

```python
df['Day_of_week']=df["Date_dt"].dt.dayofweek
df.loc[:,['Date_dt','Day_of_week']].tail(20)
```

```python
df.groupby('Day_of_week').mean() #grouping stats by day
```

```python
y=df.groupby('Day_of_week').mean()['Opening_Weekend_Gross'] 
#collecting open weeknend sales numbers
x=y.index
```

```python
print(df['Opening_Weekend_Percent_of_Gross'].mean(),df['Opening_Weekend_Percent_of_Gross'].median())
#Mean and Median of Opening weekend percent_of_gross
```

## As a percentage of gross, 28% of revenue on average comes on opening weekend so clearly opening weekend greatly affects box office success. 
## One reason for this could be a domino effect that once a movie does poorly at release, it is very hard to rebound.

```python
Weds_ratio=len(df.loc[df['Day_of_week']==2,'Day_of_week'])/len(df['Day_of_week'])
print(Weds_ratio) #ratio of movies releasing on wednesday
Fri_ratio=len(df.loc[df['Day_of_week']==4,'Day_of_week'])/len(df['Day_of_week'])
Fri_ratio #ratio of movies releasing on Friday
```

```python
fig,ax=plt.subplots(figsize=(20,15))
ax.barh(x,y,color='green')
ax.set_yticklabels(labels=['Monday',"Monday",'Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'],fontsize=20)
ax.set_xlabel('Opening Weekend Revenue ($)', fontsize=20)
ax.set_title('How The Day of the Week influences Opening Weekend Revenue',fontsize=20)
ax.tick_params(axis='both',which='major',labelsize=20)
ax.ticklabel_format(axis='x',style='plain')
fig.savefig('Day of the Week vs Opening Weekend Sales')
```

## Clearly this visualization shows that Friday and Wednesday do the best in terms of sales.
## Friday is the most common day of the week for a new release so this result is not surprising. In fact 81% of new releases were on a Friday.

## However, Wednesday did remarkbly well. Also, only 8% of new releases were on a Wednesday so there is much less competition. 

## Also by releasing a film on Wednseday, You can essentially create a "5-Day Weekend".

## Thus, I would recommend a Wednesday release.
